## Cleaning and Feature Engineering

### Target

```python
y_train = df_train["Target"]
y_test = df_test["Target"]
```

```python
df_train = df_train.drop(columns=['ID'])
df_test = df_test.drop(columns=['ID'])
ids_score = df_score["ID"]
```

```python
df_score = df_score.drop(columns=['ID'])
```

```python
all_features = []
ignored_features = [ "Target"]
num_features = []
cat_features = []
features= []




```

```python
all_features = df_train.columns.tolist()
all_features = [f for f in all_features if f not in ignored_features]
print(f"All features are {all_features}, with a length of {len(all_features)}")


```

```python
num_features = [f for f in all_features if df_train[f].dtype in ("int64","float64")]

print(f"All Numerical Features are {num_features}, with a length of {len(num_features)}")

```

```python
cat_features = [f for f in all_features if df_train[f].dtype =='object' and df_train[f].nunique() < 20]
print(f"All categorical  Features are {cat_features}, with a length of {len(cat_features)}")
```

```python
features = num_features + cat_features

skiiped_features = [f for f in all_features if f not in features]
print(f"Skipped features are {skiiped_features}, with a length of {len(skiiped_features)}")
```

### Fill Missing Values

```python
for f in num_features:
    if df_train[f].isna().sum() > 0:
        mean = df_train[f].mean()
        df_train[f] = df_train[f].fillna(mean)
        df_test[f] = df_test[f].fillna(mean)
        df_score[f] = df_score[f].fillna(mean)
```

```python
for f in cat_features:
    if df_train[f].isna().sum() > 0:
        mode = df_train[f].mode()[0]
        df_train[f] = df_train[f].fillna(mode)
        df_test[f] = df_test[f].fillna(mode)
        df_score[f] = df_score[f].fillna(mode)
```

```python
df_train[features].isna().any(axis=0), df_test[features].isna().sum(), df_score[features].isna().sum()
```

```python
### Encoding and Scaling
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
    ])
```

```python
X_train = preprocessor.fit_transform(df_train)
X_test = preprocessor.transform(df_test)
X_score = preprocessor.transform(df_score)
```
